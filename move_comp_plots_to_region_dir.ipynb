```python
import os
import shutil

# Base path where new directories should be created
base_path = '/home/users/eejap002/ica_scripts/component_plots/spatial/4_components'

# Function to process each line and perform the copy operation
def process_line(line):
    # Split the line by whitespace
    parts = line.split()
    
    # First part is the source directory
    source_dir = parts[0]
    
    # The remaining parts are the target directories
    target_dirs = parts[1:]
    
    for target_dir in target_dirs:
        # Create the full path for the target directory
        full_target_dir = os.path.join(base_path, target_dir)
        full_source_dir = os.path.join(base_path, source_dir)

        # Check if the source directory exists
        if not os.path.exists(full_source_dir):
            print(f"Source directory {source_dir} does not exist. Skipping.")
            return
        
        # Create the target directory if it doesn't exist
        if not os.path.exists(full_target_dir):
            os.makedirs(full_target_dir)
        
        # Copy contents of source_dir to full_target_dir
        for item in os.listdir(full_source_dir):
            source_path = os.path.join(full_source_dir, item)
            target_path = os.path.join(full_target_dir, item)
            if os.path.isdir(source_path):
                shutil.copytree(source_path, target_path, dirs_exist_ok=True)
            else:
                shutil.copy2(source_path, target_path)

    print(f"Copied {source_dir} to {', '.join(target_dirs)} within {base_path}")

# Main function to read the file and process each line
def main():
    # Read the input file
    with open('/gws/nopw/j04/nceo_geohazards_vol1/projects/COMET/eejap002/ica_data/all_iran_data/ica_frames.txt', 'r') as file:
        lines = file.readlines()
    
    # Process each line
    for line in lines:
        process_line(line.strip())

# Execute the main function
if __name__ == "__main__":
    main()
```
